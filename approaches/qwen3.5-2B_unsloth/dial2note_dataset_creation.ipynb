### 1. Preprocessing MedSynth dataset
    - remove duplicates
    - remove rows with missing values
    - remove too short dataset
### 2. Shuffling the dataet
### 3. Save it as "medsynth_preprocessed_shuffled"
    

```python
# Your huggingface token
your_token = ""
```

```python
from huggingface_hub import login
from datasets import load_from_disk, Dataset, load_dataset
login(token=your_token)
ds = load_dataset("Ahmad0067/MedSynth")
len(ds["train"])

# 1. Remove rows with None and missing values
ds = ds.filter(lambda x: 
    x["Dialogue"] is not None and 
    x[" Note"] is not None and 
    x["ICD10"] is not None
)

# 2. Remove rows that's too short
# They could be a noise
min_length = 50 
ds = ds.filter(lambda x: 
    len(str(x["Dialogue"]).strip()) > min_length and 
    len(str(x[" Note"]).strip()) > min_length
)

# 3. De-duplication
# based on "Dialogue"
df = ds["train"].to_pandas()
initial_count = len(df)
df = df.drop_duplicates(subset=["Dialogue"], keep="first")
print(f"Before: {initial_count}개 -> after: {len(df)}")

ds = Dataset.from_dict(df)
shuffled_ds = ds.shuffle(seed=42, keep_in_memory=True)
ds.save_to_disk("medsynth_preprocessed_shuffled")
```
